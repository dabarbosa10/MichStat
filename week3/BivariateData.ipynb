```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from scipy import stats
```

```python
data=pd.read_csv("nhanes_2015_2016.csv")
```

Scatter plot is used to visualize bivariate data. Let's see a scatter plot of arm length versus leg length. There will be a positive dependence between the two measures.

```python
plt.figure(figsize=(10,10))
sns.regplot(x="BMXLEG",y="BMXARML",data=data,fit_reg=False,scatter_kws={"alpha":0.2})
plt.grid()
```

Density plot: Darker colors indicate where a grater number of points fall.

```python
sns.jointplot(x="BMXLEG",y="BMXARML",kind='kde',data=data).annotate(stats.pearsonr)
```

```python
sns.jointplot(x="BPXSY1",y="BPXDI1",kind='kde',data=data).annotate(stats.pearsonr)
```

Now looking at two repeated measures of systolic blood pressure, taken a few minutes apart on the same person:

```python
jp=sns.jointplot(x="BPXSY1",y="BPXSY2",kind='kde',data=data).annotate(stats.pearsonr)
```
