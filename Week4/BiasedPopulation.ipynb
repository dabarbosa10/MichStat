```python
#Sampling from a biased population and not the entire population we are interested in
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
mean_oufm=155
sd_uofm=5
mean_gym=185
sd_gym=5
gymperc=0.3
totalPopSize=40000
```

```python
uofm_students=np.random.normal(mean_oufm,sd_uofm,int(totalPopSize*(1-gymperc)))
students_at_gym=np.random.normal(mean_gym,sd_gym,int(totalPopSize*(1-gymperc)))
```

```python
population=np.append(uofm_students,students_at_gym)
```

```python
plt.figure(figsize=(10,12))
plt.subplot(3,1,1)
sns.distplot(uofm_students)
plt.title("UoFM Students Only")
plt.xlim([140,200])

plt.subplot(3,1,2)
sns.distplot(students_at_gym)
plt.title("Gym Goers Only")
plt.xlim([140,200])

plt.subplot(3,1,3)
sns.distplot(population)
plt.title("Full Population")
plt.axvline(x=np.mean(population))
plt.xlim([140,200])
plt.show()
```

Sampling from entire population:

```python
numberSamp=5000
sampSize=50

mean_distribution=np.empty(numberSamp)
for i in range (numberSamp):
    random_students=np.random.choice(population,sampSize)
    mean_distribution[i]=np.mean(random_students)
    
plt.figure(figsize=(10,8))
plt.subplot(2,1,1)
sns.distplot(population)
plt.title("Full population of UofM Students")
plt.axvline(x=np.mean(population))
plt.xlim([140,200])

plt.subplot(2,1,2)
sns.distplot(mean_distribution)
plt.title("Sampling Distribution of the mean Weight of all UofM students")
plt.axvline(x=np.mean(population))
plt.axvline(x=np.mean(mean_distribution),color="black")
```

What if we take a non representative sample?

```python
numberSamps=5000
sampsize=3

mean_distribution=np.empty(numberSamps)
for i in range(numberSamps):
    random_students=np.random.choice(students_at_gym,sampSize)
    mean_distribution[i]=np.mean(random_students)
    
plt.figure(figsize=(10,8))

plt.subplot(2,1,1)
sns.distplot(population)
plt.title("FUll Population of UofM Students")
plt.axvline(x=np.mean(population))
plt.xlim([140,200])

plt.subplot(2,1,2)
sns.distplot(mean_distribution)
plt.title("Sampling Distribution of the mean weight of the gym goers")
plt.axvline(x=np.mean(population))
plt.axvline(x=np.mean(students_at_gym),color="black")
plt.xlim([140,200])

plt.show()
```
